#### Transform the necessary data

```python
def transform_money(val):
    euros = val.replace('€', '')
    if 'M' in euros:
        euros = euros.replace('M', '')
        euros = float(euros)*1000000
    elif 'K' in euros:
        euros = euros.replace('K', '')
        euros = float(euros)*1000
    return float(euros)
```

```python
data['Value_euros'] =  data['Value'].apply(transform_money)
```

```python
data['Wage_euros'] =  data['Wage'].apply(transform_money)
```

```python
data['Yearly_Wage_euros'] =  data['Wage_euros'].apply(lambda x: x*52)
```

```python
positions_dic = {'GK':'Goalkeeper','CB':'Center-back','LB':'Left back',
'RB':'Right back','RWB':'Right wing back','LWB':'Left wing back',
'RCB':'Right center back','LCB':'Left center back', 'CM' : 'Centre midfield',
'RCM' : 'Right centre midfield', 'LCM' : 'Left centre midfield', 
'LDM' : 'Left defense midfield ', 'RDM' : 'Left defense midfield',
'CDM' : 'Center defense midfield ','RAM': 'Right attacking midfield',
'LAM': 'Left attacking midfield', 'CAM': 'Centre attacking midfield',
'RM': 'Right midfield','LM': 'Left midfield','RW': 'Right wing midfield',
'LW': 'Left wing midfield','ST' : 'Stricker','RS':'Right stricker',
'LS':'Left stricker','CF':'Centre Forward','RF':'Right Forward',
'LF':'Left Forward'}

goalkeeper_pos = ['GK']
defense_pos = ['CB','LB','RB','RWB','LWB','RCB','LCB']
midfield_pos =['CM','RCM','LCM','LDM','RDM','CDM','RAM','LAM','CAM','RM','LM','RW','LW',] 
attack_pos = ['ST','RS','LS','CF','LF','RF']

        
```

```python
position_type = []
long_position = []
for i in range(len(data)):
    if data['Position'].iloc[i] in goalkeeper_pos:
        position_type.append('Goalkeeper')
    elif data['Position'].iloc[i] in defense_pos:
        position_type.append('Defense')
    elif data['Position'].iloc[i] in midfield_pos:
        position_type.append('Midfield')        
    elif data['Position'].iloc[i] in attack_pos:
        position_type.append('Attack')        
    else:
        position_type.append('Unknonkn') 
    long_position.append(positions_dic.get(data['Position'].iloc[i]))
        
```

```python
data['PositionType'] = position_type
data['PositionLong'] = long_position
```

### Heatmap correlation of all the the values

```python
sns.set(style="white")

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(14, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.set(font_scale=1)

ax = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()
```
